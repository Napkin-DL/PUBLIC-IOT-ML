```python
import random
import glob

image_list = glob.glob('./samples/*/*')
random.shuffle(image_list)

label_name = ''
pre_label_name = []
for img_path in image_list:
    images = img_path.split('/')
    label_name = images[2]
    text = '<h2 style="color:black;font-size:20px;"><strong>'+label_name+'</strong></h2>\n<p><img width="500" alt="'+label_name+'" src="'+img_path+'"></p><br><br>'
    if label_name not in pre_label_name:
        print(text)
        pre_label_name.append(label_name)
```
